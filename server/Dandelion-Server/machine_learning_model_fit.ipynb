```python
from sklearn.model_selection import train_test_split
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
# import tensorflow as tf
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import glob

```

```python
# 이미지 경로 지정
data_dir = '/root/Dandelion/Dataset_append'
```

```python
# 카테고리 지정
categories = ['Pepero_Amond', 'Pepero_Crunch', 'Pepero_Original']
nb_class = len(categories)
```

```python
# 이미지 사이즈 지정
image_w = 224
image_h = 224
pixels = image_w * image_h * 3
```

```python
# 이미지마다 원핫 인코딩
x = []
y = []

# 폴더 안 데이터 다 가져오기
for idx, c in enumerate(categories):
    label = [0 for i in range(nb_class)]
    label[idx] = 1
    image_dir = data_dir +"/"+c
    files = glob.glob(image_dir+'/*.jpg')
    # print(files)
    # 파일 안에 있는 이미지 오픈해서 어레이 값으로 변경
    for i, f in enumerate(files):
        print(f)
        img = Image.open(f)
        img = img.convert("RGB")
        img = img.resize((image_w, image_h))
        data = np.asarray(img)
        x.append(data)
        y.append(label)
    
```

```python
# 실제 array 형태 바꾸기 / 형태 확인
x = np.array(x)
x.shape
```

```python
y = np.array(y)
y.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2)
```

```python
# train / test 데이터셋 나누기
image_data = (X_train, X_test, Y_train, Y_test)
np.save('./image_data.npy', image_data)
```

```python
# 표준화 / 정규화 사용
X_train = X_train.astype('float32')/255
X_test = X_test.astype('float32')/255
```

```python
# <입력층>
# 입력 변수, 출력 변수 만들기
# 몇개가 들어올지 모르니 None으로 표기
x = tf.placeholder(tf.float32, [None, 224, 224, 3])
y = tf.placeholder(tf.float32, [None, 3])
# 해당 코드는 Tensorflow 1.0 코드
```

```python
# <은닉층>
# 필터 크기 = [필터 가로, 필터 세로, 색 수, 채널]
# tf.random_normal : 정규 분포에 다른 숫자를 난수 값으로 표현
w1 = tf.Variable(tf.random_normal([3,3,3,32], stddev=0.01))
```

```python
L1 = tf.nn.conv2d(x,w1, strides=[1,1,1,1], padding='SAME')
L1
```

```python
# ksize=[1,2,2,1] : pooling 할 때 사용하는 필터 2*2 사이즈
# maxpool을 사용하더라도 이미지 크기 같게 하려면 stride 1, padding = 'SAME'
L1 = tf.nn.relu(L1)
L1 = tf.nn.max_pool(L1, ksize=[1,2,2,1], strides=[1,2,2,1], padding = 'SAME')
L1
```

```python
# <은닉층 -2>
# 필터 : [가로, 세로, 앞단의 채널, 채널]
# 필터 크기 3*3, 스트라이드 1, 패딩 same
w2 = tf.Variable(tf.random_normal([3,3,32,64], stddev=0.01))
L2 = tf.nn.conv2d(L1, w2, strides=[1,1,1,1], padding='SAME')
L2
```

```python
# relu사용, max pooling 사용
L2 = tf.nn.relu(L2)
L2 = tf.nn.max_pool(L2, ksize=[1,2,2,1], strides=[1,2,2,1], padding = 'SAME')
L2
```

```python
# <은닉층 -3>
# 필터크기 3*3, 스트라이드1, 패딩 same
w3 = tf.Variable(tf.random_normal([3,3,64,64], stddev=0.01))
L3 = tf.nn.conv2d(L2, w3, strides=[1,1,1,1], padding='SAME')
L3
```

```python
#relu 사용, max pooling 사용
L3 = tf.nn.relu(L3)
L3 = tf.nn.max_pool(L3, ksize=[1,2,2,1], strides=[1,2,2,1], padding = 'SAME')
L3
```

```python
# <Fully connected 층>
# 여기서 w4는 filter가 아닌 weight 값으로 생각
# w4 = [앞단의 크기, 채널]
w4 = tf.Variable(tf.random_normal([28*28*64, 256], stddev=0.01))
w4
```

```python
# [행의 수, 열의 수] : -1은 모르겠다는 뜻
L4 = tf.reshape(L3, [-1, 28*28*64])
L4
```

```python
# L4와 w4을 행렬의 곱하고 relu 적용
L4 = tf.nn.relu(tf.matmul(L4, w4))
L4
```

```python
# <출력층>
# [입력 개수, 출력 개수(클래스 개수)]
w5 = tf.Variable(tf.random_normal([256, 3], stddev=0.01))
w5
```

```python
# L4와 w5 행렬의 곱
model = tf.matmul(L4, w5)
model
```

```python
# softmax로 확률 계산
hypothesis = tf.nn.softmax(model)
```

```python
# softmax로 확률값 계산 / cross entropy로 손실값 계산
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=model, labels=y))
```

```python
# AdamOptimizer은 내부적 모멘텀을 이용해 학습률 조절
optimizer = tf.train.AdamOptimizer(0.001).minimize(cost)
```

```python
### 튜닝
#세션 열고 변수 초기화
saver = tf.train.Saver()
sess = tf.Session()
sess.run(tf.global_variables_initializer())
saver.save(sess, './my_model/my_test_model', global_step=0)
```

```python
# 데이터셋이 많은 경우 미니배치를 사용하여 표본추출
batch_size = 100
for epoch in range(1,11):
    avg_cost = 0
    # 배치사이즈만큼 뽑아서 사용
    for i in range(int(np.ceil(len(X_train)/batch_size))):
        x_ = X_train[batch_size*i : batch_size*(i+1)]
        y_ = Y_train[batch_size*i : batch_size*(i+1)]
        _,cost_val = sess.run([optimizer,cost], feed_dict={x:x_,y:y_})
        avg_cost += cost_val
    print('Epoch:','%04d'%(epoch),
           'cost: ','{:.9f}'.format(avg_cost/len(X_train)))
```

```python
# 정확도 계산
# tf.argmax(X,n) : X 배열에서 가장 큰 값을 ‘열(1)’ 기준 몇번째인지 index를 반환
# tf.cast() : 입력한 값의 결과를 지정한 자료형으로 변환해줌
# tf.equal() : x, y를 비교하여 boolean 값을 반환

is_correct = tf.equal(tf.argmax(hypothesis,1),tf.argmax(y,1))
accuracy=tf.reduce_mean(tf.cast(is_correct,tf.float32))
print("정확도 : ",sess.run(accuracy,feed_dict={x:X_test,y:Y_test}))
```

```python
from tensorflow import keras
from sklearn.model_selection import train_test_split
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import glob
```

```python
# 이미지 경로 지정
data_dir = '/root/Dandelion/Dataset_append'
```

```python
# 카테고리 지정
categories = ['Pepero_Amond', 'Pepero_Crunch', 'Pepero_Original', 'Pocky_Original', 'Pocky_Strawberry', 'Pocky_Thin']
nb_class = len(categories)
```

```python
# 이미지 사이즈 지정
image_w = 224
image_h = 224
pixels = image_w * image_h * 3
```

```python
# 이미지마다 라벨 번호 지정(아몬드 : 1, 크런치 : 2, 오리지널 : 3)
x = []
y = []

# 폴더 안 데이터 다 가져오기
for idx, c in enumerate(categories):
    label = [0 for i in range(nb_class)]
    label[idx] = 1
    image_dir = data_dir +"/"+c
    files = glob.glob(image_dir+'/*.jpg')
    # print(files)
    # 파일 안에 있는 이미지 오픈해서 어레이 값으로 변경
    for i, f in enumerate(files):
        print(f)
        img = Image.open(f)
        img = img.convert("RGB")
        img = img.resize((image_w, image_h))
        data = np.asarray(img)
        x.append(data)
        y.append(label)
    
```

```python
x
```

```python
y
```

```python
# 실제 array 형태 바꾸기 / 형태 확인
x = np.array(x)
x.shape
```

```python
y = np.array(y)
y.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
```

```python
# train / test 데이터셋 나누기
image_data = (X_train, X_test, Y_train, Y_test)
np.save('./image_data_tf.npy', image_data)
```

```python
# 저장한 어레이값 불러와서 사용 / allow_pickle 사용
X_train, X_test, Y_train, Y_test =\
np.load('./image_data_tf.npy', allow_pickle = True)
X_train.shape
```

```python
X_train_scaled = X_train / 255.0
```

```python
X_train_scaled.shape
```

```python
X_test_scaled = X_test / 255.0
```

```python
X_test_scaled.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
model = keras.Sequential()
model.add(keras.layers.Conv2D(32, kernel_size=3, activation = 'relu', padding = 'same', input_shape=X_train_scaled.shape[1:]))
```

```python
model.add(keras.layers.MaxPooling2D(2))
```

```python
model.add(keras.layers.Conv2D(64, kernel_size=3, activation='relu', padding='same'))
model.add(keras.layers.MaxPooling2D(2))
```

```python
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(128, activation='relu'))
# model.add(keras.layers.Dropout(0.4))
model.add(keras.layers.Dense(6, activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics='accuracy')
checkpoint_cb=keras.callbacks.ModelCheckpoint('best-cnn-model.h5')
early_stopping_cb=keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)
history = model.fit(X_train_scaled, Y_train, epochs=20, validation_data = (X_test_scaled, Y_test), callbacks = [checkpoint_cb, early_stopping_cb])
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train','val'])
plt.show()
```

```python
model.evaluate(X_test_scaled, Y_test)
```
